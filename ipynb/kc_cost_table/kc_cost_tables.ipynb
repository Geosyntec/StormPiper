```python
import pandas
```

```python
df = (
    pandas.read_excel("./kc_cost_tables_2024WQBE.xlsx", sheet_name=0)
    .rename(columns=lambda c: c.replace(" ", "_").lower())
    .assign(
        description=lambda df: "WQBE "
        + df.code
        + " "
        + df.description.str.split("_").str[-1]
    )
)
df.head()
```

```python
cols = [
    "description",
    "code",
    "tacoma_type",
    "sizing_method",
    "sizing_unit",
    "tdcc_cpu",
    "tincc_cpu",
    "totalcapital_cpu",
    "om_cpu",
]
```

```python
df[cols].round(2).reset_index().to_json(
    "kc_cost_table.json", orient="records", indent=2
)
```
